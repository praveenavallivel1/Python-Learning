```python
import multiprocessing
import time
from random import randint

def sumUp():
    i = 0
    while i <=30:
        print('Summing Up Values:'.format(i))
        time.sleep(randint(1,3))
        i += 1
def sumDown():
    i = 30
    while i >= 0:
        print('Summing Down Values:'.format(i))
        time.sleep(randint(1,3))
        i -= 1       
if __name__ == '__main__':
    
    # initialize the parallel worker set up
    worker1 = multiprocessing.Process(target=sumUp)
    worker2 = multiprocessing.Process(target=sumDown)
    
    # invoke the workers
    worker1.start()
    worker2.start()
    
    # combine the results
    worker1.join()
    worker2.join()
```

```python
worker1.terminate()
```
