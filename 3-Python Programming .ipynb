```python
# placeholder---using for loop
start_time = time.time()
squares = []

for num in range(1,100000001):
    new_num = num**2
    squares.append(new_num)
for square in squares[-10:]:
    print(square)

print(time.time() - start_time)
```

```python
# placeholder---using list comprehension
start_time = time.time()
squares = [num**2 for num in range(1,100000001)]

for square in squares[-10:]:
    print(square)

print(time.time() - start_time)
```

```python
#Handling Missing Values
import pandas
import numpy as np
data = {'one': pandas.Series([1, 2, 5], index=['a', 'b', 'e']),
        'two': pandas.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}
table = pandas.DataFrame(data)
print("Before:\n",table)

table['one'] = table['one'].replace(np.nan, -99)
table['two'].fillna(-99,inplace = True)
print("After:\n",table)
```

```python
#Handling Missing values - imputer
import pandas
import numpy as np
data = {'one': pandas.Series([1, 2, 5], 
               index=['a', 'b', 'e']),
        'two': pandas.Series([1, 2, 3, 4], 
               index=['a', 'b', 'c', 'd'])}
table = pandas.DataFrame(data)
print("Before:\n",table)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(table)
imputed_data = imputer.transform(table.values)
print("After:\n",pandas.DataFrame(imputed_data, columns=['one','two']))
```

```python
#Labelencoder
import pandas as pd
data = {'Customer_id': pd.Series([1,2,3,4,5]),
        'Loan_type': pd.Series(['Home Loan',
                                'Personal Loan',
                                'Education Loan',
                                'Home Loan',
                                'Credit Loan']),
        'Income': pd.Series(['30K','25K','15K',
                             '40K','35K'])}
loan_info = pd.DataFrame(data)
print(loan_info)

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

loan_info_upd=loan_info.apply(labelencoder.fit_transform)
print(loan_info_upd)
```

```python
#import the library used to query a website
import requests

import pandas as pd
 
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup

#specify the url
quotes = "https://www.goodreads.com/quotes"

#Query the website/url and return the html to the variable 'page'
page = requests.get(quotes)
 
#Fetching the contents of the website as a document
soup = BeautifulSoup(page.text, 'html.parser')

#Find all the lines in the document whose class was quoteDetails which has all
#the details of the quotes - Authorname, quote
quote_text=soup.find_all(class_="quoteDetails")

#Creating an empty dataframe to store the retrieved details
df1=pd.DataFrame() 

#Loop to fetch each of the category- Authorname,quote separately and prints them.
for text1 in quote_text:
      
      #Fetch the author name when it finds the class as authorOrTitle.
      #Fetch the quote when it finds the class as quoteText.
      #Fetch the likes when it finds the class as right.
      likes=text1.find(class_="right")
      num_like=likes.find('a')
      for num_likenew in num_like:
          author_name=text1.find_all(class_="authorOrTitle")
          quotetext=text1.find_all(class_="quoteText")
          for qu in quotetext:
                for names in author_name:
                    #Appending the data to the dataframe
                    df1 = df1.append({'Authorname': names.contents[0].replace('\n',''),
                                       "Quote" : qu.contents[0].replace('\n',''),
                                       "Likes": num_likenew },
                                        ignore_index=True)

print(df1.head())
#Storing the data to a csv file
df1.to_csv("Extracted Quotes.csv")          
```
